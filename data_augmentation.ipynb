```python
from PIL import Image
from skimage import io, transform
from skimage.color import rgb2gray
```

```python
import numpy as np
import os
```

```python
for path, subdirs, files in os.walk('./data/train/org'):
    for i in range(len(files)):
        org_path = './data/train/org/'+files[i]
        gt_path = './data/train/gt/'+files[i]
        for degree in range(60,360,60):
            img1 = rgb2gray(io.imread(org_path))
            img2 = io.imread(gt_path)
            img1 = transform.rotate(img1, degree)
            img2 = transform.rotate(img2, degree)
            io.imsave(org_path[:-4]+'('+str(degree)+')'+'.tif',img1)
            io.imsave(gt_path[:-4]+'('+str(degree)+')'+'.tif',img2)
```
